```python
import pandas as pd

dados = pd.read_csv("dados_tratados.csv")

dados.head()
```

```python
dados_processados = dados.drop('customerID', axis=1)

dados_processados = pd.get_dummies(dados_processados, drop_first=True)

print("Dimensões do DataFrame após o pré-processamento:", dados_processados.shape)
display(dados_processados.head())
```

DataFrame dados_processados agora está no formato para ser usado por algoritmos de Machine Learning: totalmente numérico.

Passamos de 21 colunas para 31, pois as variáveis de texto foram transformadas em colunas binárias.

**Separação dos Dados em Treino e Teste**

DataFrame em duas partes:
X: Todas as colunas que serão usadas como "pistas" para a previsão (as features).

y: A coluna que queremos prever (Churn).

```python
from sklearn.model_selection import train_test_split

# Separar as features (X) e o alvo (y)
X = dados_processados.drop('Churn', axis=1)
y = dados_processados['Churn']

# Dividir os dados em conjuntos de treino e teste

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# VERIFICAÇÃO: Vamos verificar as dimensões dos conjuntos criados
print("Dimensões de X_train:", X_train.shape)
print("Dimensões de X_test:", X_test.shape)
print("Dimensões de y_train:", y_train.shape)
print("Dimensões de y_test:", y_test.shape)
```

Dimensões de X_train: (4930, 30): 4930 clientes para treinar modelo, e para cada um, temos 30 características (colunas) para usar como "pistas".

Dimensões de X_test: (2113, 30): 2113 clientes para o teste final. Note que o número de colunas (30) é o mesmo, o que é crucial.

Dimensões de y_train: (4930,): Para cada um dos 4930 clientes de treino, temos a resposta correta (se ele evadiu ou não).

Dimensões de y_test: (2113,): E para os 2113 clientes de teste, também tem o gabarito, para comparar com as previsões do modelo.

A soma 4930 + 2113 = 7043, que é o número total de clientes.

A divisão 70/30 foi feita perfeitamente.

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression(max_iter=1000)

modelo.fit(X_train, y_train)

print("Modelo treinado com sucesso!")
```

**Todas as variáveis na mesma escala**

A técnica mais comum é o StandardScaler, que transforma cada coluna para que ela tenha média 0 e desvio padrão 1.

**inserir** o escalonamento no fluxo, é um passo de pré-processamento que deve vir depois da divisão treino/teste.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

modelo = LogisticRegression()

modelo.fit(X_train_scaled, y_train)

print("Modelo treinado com sucesso e sem avisos!")
```

Para avaliar a performance, usarei três métricas importantes:

Acurácia (Accuracy): A porcentagem de previsões que o modelo acertou no total. É a métrica mais simples.

Matriz de Confusão (Confusion Matrix): Uma tabela que nos mostra em detalhes os tipos de acertos e erros do modelo (ex: previu que o cliente sairia e ele realmente saiu).

Relatório de Classificação (Classification Report): Fornece métricas mais detalhadas como precisão e recall, que são muito úteis para entender a performance do modelo para cada classe (Churn "Sim" e Churn "Não").

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = modelo.predict(X_test_scaled)

# Calcular a acurácia
acuracia = accuracy_score(y_test, y_pred)
print(f"Acurácia do modelo: {acuracia * 100:.2f}%")

print("\n--------------------------------------------------------\n")

print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))

print("\n--------------------------------------------------------\n")

# Exibir o Relatório de Classificação
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred, target_names=['Não Churn (0)', 'Churn (1)']))
```

**Relatório de Classificação**

O relatório transforma os números da matriz de confusão em métricas percentuais, que são mais fáceis de comparar.
Focando na linha mais importante para o negócio, que é a Churn (1):

Precision (0.67): De todos os clientes que o modelo alertou como "potencial Churn", 67% deles realmente cancelaram. Mostra que os "alarmes" do modelo são razoavelmente confiáveis.

Recall (0.57): métrica mais crítica. De todos os clientes que realmente cancelaram, o modelo conseguiu identificar 57% deles.

F1-score (0.61): média harmônica entre a precisão e o recall. Serve como uma métrica única que equilibra os dois.

**Conclusão e PróximosPassos**

O primeiro modelo de Regressão Logística é bom em prever os clientes que vão ficar (recall de 90%), mas tem uma performance moderada para prever os que vão sair (recall de 57%).

Ele consegue identificar mais da metade dos clientes que estão em risco, o que já é extremamente valioso para a empresa.

Onde podem melhorar? O principal ponto é aumentar o recall para a classe "Churn (1)". Queremos "deixar passar" o menor número possível de clientes que estão prestes a cancelar, mesmo que isso signifique aumentar um pouco o número de falsos positivos.
